```python
section_types = []

fulltext = {}

for file in datafiles:
    pmcid = file.stem
    pub = Publication(pmcid)
    pub.load_json()
    pub.parse_sections()
    section_types.extend(pub.sections.keys())
    pub.combine_text()
    fulltext[pmcid] = pub.fulltext
```

```python
system_msg = """
You are an expert in psychology and neuroscience.
You should be as specific and as comprehensive as possible in your responses.
Your response should be a JSON object with no additional text.
"""

client = ChatClientFactory.create_client("anthropic", api_key, system_msg=system_msg)
```

```python
text = fulltext['10908822']
prompt = f"""
The following is the full text from a scientific publication in the field of 
psychology and neuroscience.  In this field, researchers study specific psychological
constructs, which are the building blocks of the mind, such as memory, attention, 
theory of mind, and so on.  To study them, researchers use experimental tasks or 
surveys.

Your job is to identify which psychological concepts and tasks or surveys are used in 
the following text.  Please return the results in JSON format, with the keys being 'construct'
and 'task' and the values being lists of strings that represent the concepts and tasks that are 
present in the text.  Be as specific as possible, giving the specific names of the tasks or surveys.
Respond on with JSON, without any additional text or description.

{text}
"""

response = client.chat(prompt)
```
