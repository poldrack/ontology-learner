```python
import json
import pandas as pd
from pathlib import Path
from dotenv import load_dotenv
import os
from ontology_learner.publication import Publication

load_dotenv()
datadir = Path(os.getenv('DATADIR'))
jsondir = datadir / 'json'
```

```python
coord_df = pd.read_csv(datadir / 'coordinate_extraction/coords_df_gpt4.csv')

pmcids = sorted(list(coord_df.pmid.unique()))
pmcid_to_pmid = {}
bad_pmcids = []
for pmcid in pmcids:
    fname = jsondir / f'{pmcid}.json'
    p = Publication(pmcid, datadir=jsondir)
    p.load_json()
    try:
        pmid = p.data['documents'][0]['passages'][0]['infons']['article-id_pmid']
    except KeyError:
        print(f'No pmid found for {pmcid}')
        bad_pmcids.append(pmcid)

    pmcid_to_pmid[str(pmcid)] = str(pmid)
    
```

```python
with open(datadir / 'pmcid_to_pmid.json', 'w') as f:
    json.dump(pmcid_to_pmid, f, indent=4)
```
