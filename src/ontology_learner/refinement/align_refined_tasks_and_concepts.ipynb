### Load refined tasks and concepts

```python
refined_tasks_raw = load_jsonl(datadir / 'task_refinement_results' / 'batch_67464c2189dc8190a9ecc642a8d7afd9.jsonl')
refined_task_dict = parse_jsonl_results(refined_tasks_raw)

orig_tasks_dir = datadir / 'task_results'
orig_task_dict = load_original_results(orig_tasks_dir)

refined_concepts_raw = load_jsonl(datadir / 'construct_refinement_results' / 'batch_67465651786c8190b5655afa8284ab49.jsonl')
refined_concept_dict = parse_jsonl_results(refined_concepts_raw)

orig_concepts_dir = datadir / 'concept_results'
orig_concept_dict = load_original_results(orig_concepts_dir)
```

Pull out individual concepts and tasks from refined lists



```python
expanded_task_dict = {}

for clusternum, clusterdict in refined_task_dict.items():
    for taskname, taskdict in clusterdict.items():
        if taskname not in orig_task_dict:
            expanded_task_dict[taskname] = taskdict
        else:
            print(f'{taskname} already in original task dict')

print(f'found {len(expanded_task_dict)} refined tasks')
```

Using expanded_task_dict, integrate all of the info from the original task dict

```python

missing_ids = []
for taskname, taskdict in expanded_task_dict.items():
    for label in taskdict['labels']:
        custom_id = 'task-' + label.lower().replace(' ', '_')
        if custom_id in orig_task_dict:
            if 'original_task_ids' not in taskdict:
                expanded_task_dict[taskname]['original_task_ids'] = [custom_id]
            else:
                expanded_task_dict[taskname]['original_task_ids'].append(custom_id)
        else:
            missing_ids.append(custom_id)
            print(f'{custom_id} not in original task dict')

print(f'found {len(missing_ids)} missing task ids')
print(f'found {len(expanded_task_dict)} refined tasks')
```
